```python
import numpy as np

def generate_data(m=10):
    np.random.seed(42)
    x0 = np.ones((m,1))  # Columna de unos
    x1 = np.random.rand(m,1)  # Primera característica aleatoria
    x2 = np.random.rand(m,1)  # Segunda característica aleatoria
    x = np.hstack((x0, x1, x2))
    y = 1 + 2 * x1 + 3 * x2
    return x, y

def generate_matriz(array):
    m = len(array)  # Obtener la longitud del array de entrada
    x0 = np.ones((m, 1))  # Columna de unos
    x1 = np.array(array).reshape(-1, 1)  # Convertir el array de entrada en una columna de la matriz x
    x = np.hstack((x0, x1))
    y = 1 + 2 * np.array(array)  
    return x, y


def add_noise(y, percent_noise=5):
    noise = percent_noise / 100 * (np.random.randn(len(y), 1)/1000)
    y_noisy = y + noise
    #print("y: ", len(y))
    return y_noisy

def calcular_hat(x, y):
    w_star = np.linalg.inv(x.T @ x) @ x.T @ y
    y_hat = x @ w_star
    return y_hat

def calcular_mse(a, b):
    #print("a: ", len(a))
    return np.square(np.subtract(a, b)).mean()

def main():
    x, y = generate_data()
    
    y_hat = calcular_hat(x, y)
    print("y_hat = ", y_hat)
    mse_y = calcular_mse(y, y_hat)
    print("MSE entre y y y_hat: ", mse_y,"\n")
    
    
    W_x, W_y = generate_matriz([1, 2, 3])
    y_noisy = add_noise(y)
    w_hat = calcular_hat(x, y_noisy)
    print("w_hat = ", w_hat)
    mse_w = calcular_mse(W_y,  w_hat)
    print("MSE entre W_y y w_hat: ", mse_w, "\n")

if __name__ == "__main__":
    main()
```
