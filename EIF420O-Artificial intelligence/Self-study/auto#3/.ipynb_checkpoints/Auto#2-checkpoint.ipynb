```python
from sympy import symbols, diff, Matrix
```

```python
x, y= symbols('x y')
```

```python
f = 3*x**2 - y**2 + 1
```

```python
# Calcular las derivadas parciales de segundo orden
f_xx = diff(f, x, x)
f_xy = diff(f, x, y)
f_yx = diff(f, y, x)
f_yy = diff(f, y, y)
```

```python
# Construir la matriz Hessiana
Hessian_matrix = Matrix([[f_xx, f_xy], [f_yx, f_yy]])
```

```python
# Definir el punto crítico
critical_point = {x: 0, y: 0}
```

```python
# Evaluar la matriz Hessiana en el punto crítico
Hessian_matrix_evaluated = Hessian_matrix.subs(critical_point)
```

```python
# Calcular el determinante y la traza de la matriz Hessiana
determinant = Hessian_matrix_evaluated.det()
trace = Hessian_matrix_evaluated.trace()
```

```python
if determinant < 0:
    print("El punto crítico es un punto de silla.")
else:
    print("El punto crítico no es un punto de silla.")
```
