```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def calcular_modelo(w_0, w_1, x):
    return w_0 + w_1 * x

def calcular_MSE(y, y_pred):
    '''Calcula el error cuadrático medio entre el dato original (y)
       y el dato generado por el modelo (y_pred)'''
    N = y.shape[0]
    error = np.sum((y - y_pred) ** 2) / N
    return error

def gradiente_descendente(w_0, w_1, alpha, x, y):
    '''Algoritmo del gradiente descendente para minimizar el error
       cuadrático medio'''
    N = x.shape[0]  # Cantidad de datos

    # Gradientes: derivadas de la función de error con respecto
    # a los parámetros "w" y "b"
    dw_0 = -(2 / N) * np.sum(y - (w_0 + w_1 * x))
    dw_1 = -(2 / N) * np.sum(x * (y - (w_0 + w_1 * x)))

    # Actualizar los pesos usando la fórmula del gradiente descendente
    w_0_nuevo = w_0 - alpha * dw_0
    w_1_nuevo = w_1 - alpha * dw_1

    return w_0_nuevo, w_1_nuevo

# Datos de ejemplo
np.random.seed(0)
x = 2 * np.random.rand(50, 1)
y = 3 + 4 * X + np.random.randn(50, 1)

# Inicialización de parámetros
w_0 = 0
w_1 = 0
alpha = 0.01  # Tasa de aprendizaje
iteraciones = 1000

# Bucle de entrenamiento
for i in range(iteraciones):
    w_0, w_1 = gradiente_descendente(w_0, w_1, alpha, x, y)
    if i % 100 == 0:
        y_pred = calcular_modelo(w_0, w_1, x)
        mse = calcular_MSE(y, y_pred)
        print(f"Iteración {i}: w_0 = {w_0}, w_1 = {w_1}, MSE = {mse}")

# Predicciones finales y visualización
y_pred = calcular_modelo(w_0, w_1, x)
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', label='Línea de regresión')
plt.legend()
plt.show()
```
