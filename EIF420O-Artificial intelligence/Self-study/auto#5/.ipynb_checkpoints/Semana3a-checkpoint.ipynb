```python
import random, math
random.seed(666)

def f(x, y):
    return 2*x**2*y**2 - 6*x*y**2 + 3*y**2 - 1
def g(x, y):
    return (x - 2)**2 + 2*(y-1)**2 + x
def f_quiz(x,y):
    return (x - 1/2)**2 + (y - 1/4)**2 + 3
```

```python
def gradient_f(x,y):
    f_x = (4*x - 6)*y**2
    f_y = (4*x**2 - 12*x + 6)*y
    return f_x,f_y
def gradient_g(x,y):
    g_x = 2*x - 3 #error en la derivada
    g_y = 4*(y - 1)
    return g_x, g_y
def gradient_f_quiz(x,y):
    f_qx = 2*x + 1
    f_qy = 2*y - 1/2
    return f_qx, f_qy
```

```python
def gradient_descent(gradient, critical_point, f, 
                     max_iter=100, epsilon =1e-5, 
                     initial = None, 
                     lr=0.01): #lr = Learning rate
    if initial is None:
        initial = random.random(), random.random()
    actual_min = f(*critical_point) # * es igual a spred de javascript, osea desempaca
    xn, yn = initial
   
    for iter in range(max_iter):
        grad_x, grad_y = gradient(xn, yn)
        xn, yn = (xn - lr * grad_x, yn - lr * grad_y)
        estimated_min = f(xn, yn)
        print(f"{iter=:2} aprox_{iter} = ({xn=:},{yn=:}) {actual_min} ==? {estimated_min}")
        if abs(actual_min - estimated_min) < epsilon:
            print(f"*** convergence achieved *** at {iter=:}")
            break
        
```

```python
gradient_descent(gradient_f_quiz, (0.5,0.25), f_quiz, max_iter = 100, lr = 0.025)
```
